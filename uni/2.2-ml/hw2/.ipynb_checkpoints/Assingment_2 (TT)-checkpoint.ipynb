````python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("task 1.csv")

# Set lebels (user_id = 0 -> 1, user_id != 0 -> 0)
for index, value in df["user_id"].items():
    if value == 0:
        df["user_id"][index] = 1
    else:
        df["user_id"][index] = 0

y = df["user_id"]

X = df.drop(columns=["index", "user_id", "time", "date", "sites"])
X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=123, stratify=y)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.25, random_state=123, stratify=y_train)
````

````python
from keras.models import Sequential
from keras.layers import Dense, Dropout

import keras.backend as K

# metrics for calculating f1 score used it model training
def recall(y_true, y_pred):
    tp = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    pos = K.sum(K.round(K.clip(y_true, 0, 1)))
    return tp / (pos + K.epsilon())

def precision(y_true, y_pred):
    tp = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    pred = K.sum(K.round(K.clip(y_pred, 0, 1)))
    return tp / (pred + K.epsilon())

def f1(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2*((p*r)/(p+r+K.epsilon()))



# Define function to create model 
def create_model(neurons=16, num_layers=1, dropout = 0.0):
    model = Sequential()
    model.add(Dense(neurons, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dropout(dropout))
    for i in range(num_layers):
        model.add(Dense(neurons, activation='relu'))
        model.add(Dropout(dropout))
    model.add(Dense(1, activation='sigmoid'))
    # monitor f1 as dataset is unbalanced, also take into consideration precision, recall, accuracy - try to maximize all of them
    model.compile(optimizer="adam", loss="binary_crossentropy",
              metrics=[f1, precision, recall, 'accuracy'])
    return model
````

````python
from keras.callbacks import EarlyStopping

es = EarlyStopping(min_delta=0.001, restore_best_weights= True) # Add early stopping

model = create_model()

# Train model without any fine tuning

model.fit(X_train, y_train, 
          callbacks=[es],
          validation_data=(X_val, y_val),
          batch_size = 256,
          epochs = 1000,
          verbose = 1
           )
````

````python
from sklearn.metrics import classification_report

# evaluate the model without fine tuning
y_pred = model.predict(X_test, verbose=0)
y_pred = (y_pred > 0.5)  
print(classification_report(y_test, y_pred))
````

````python
from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier

model = KerasClassifier(build_fn=create_model, verbose=1)

param_grid = {
    'neurons': [32, 64, 128],
    'num_layers': [1, 2, 3],
    'dropout' : [0.0, 0.1, 0.2, 0.3]
}

# Fine tune model
grid_search = GridSearchCV(model, param_grid=param_grid, cv=3, scoring='f1', verbose=0)
grid_result = grid_search.fit(X_train, y_train, verbose=0)

print("Best f1 score: {:.2f} using {}".format(grid_result.best_score_, grid_result.best_params_))
````

````python

best_params = grid_search.best_params_

es = EarlyStopping(min_delta=0.001, restore_best_weights= True)

# Create a new model with the fine tuned hyperparameters
model = create_model(neurons=best_params['neurons'], num_layers=best_params['num_layers'], dropout=best_params['dropout'])

model.fit(X_train, y_train, epochs=100, batch_size=256, validation_data=(X_val, y_val), callbacks=[es])
````

````python
from sklearn.metrics import classification_report

# evaluate the model on some test data
y_pred = model.predict(X_test, verbose=0)
y_pred = (y_pred > 0.5)
print(classification_report(y_test, y_pred))
````

````python
import numpy as np
import cv2
import os
from keras.utils.np_utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator

def preprocess_image(img_path, img_size):
    # Convert to bmp if filename doesn't have .bmp extension
    if not img_path.endswith('.bmp'):
        new_img_path = img_path + '.bmp'
        cv2.imwrite(new_img_path, cv2.imread(img_path))
        img_path = new_img_path
    
    # Load image and label from file
    img = cv2.imread(img_path)
    label = int(os.path.basename(img_path).split('_')[1].split('.')[0])

    # Crop borders as they have black color which can become noiыe in model training
    top_border = int(0.05 * img.shape[0])  # 5% of height
    bottom_border = int(0.05 * img.shape[0])
    left_border = int(0.05 * img.shape[1])  # 5% of width
    right_border = int(0.05 * img.shape[1])
    
    img = img[top_border:img.shape[0]-bottom_border, 
                        left_border:img.shape[1]-right_border]

    # Resize image to fixed size - size of the smallest picture
    img = cv2.resize(img, img_size)

    # Convert image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Normalize pixel values in the range of 0-1
    gray_img = gray_img.astype('float32') / 255.0

    return (gray_img[:,:,np.newaxis], label)

# function for creating datasets
def data_loader(path, img_size=(103, 96)):
    data = []
    for filename in os.listdir(path):
        img_path = os.path.join(path, filename)
        img, label = preprocess_image(img_path, img_size)
        data.append((img, label))
    return data

# Size of the smallest picture in datasets
img_size = (103, 96)


# create dataset for training
train_path = 'train/'
train_data = data_loader(train_path)

# create dataset for validation
train_data, val_data = train_test_split(train_data, test_size=0.2, stratify=[data[1] for data in train_data])

# create batch genenetors to minimize RAM usage
datagen = ImageDataGenerator()
# Set up data generator for train data
train_generator = datagen.flow(
        x=np.array([data[0] for data in train_data]),
        y=to_categorical(np.array([data[1] for data in train_data]), num_classes=10),
        batch_size=128
)

# Set up data generator for validation data
val_generator = datagen.flow(
        x=np.array([data[0] for data in val_data]),
        y=to_categorical(np.array([data[1] for data in val_data]), num_classes=10),
        batch_size=128
)

# Set up data generator for test data
test_path = 'content/test/'
test_data = data_loader(test_path)
test_generator = datagen.flow(
        x=np.array([data[0] for data in test_data]),
        y=to_categorical(np.array([data[1] for data in test_data]), num_classes=10),
        batch_size=128
)
````

````python
# model architecture was taken from https://www.kaggle.com/code/codeblogger/convolutional-neural-network-cnn-tutorial
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, MaxPooling2D

# function for creating model

def create_model(filters=25, neurons=128, dropout=0.1):
    
    model = Sequential()

    model.add(Conv2D(filters=filters, kernel_size=(3,3), strides=1, padding='same', activation='relu', input_shape=(96,103,1)))
    model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='same'))

    model.add(Conv2D(filters=int(filters*0.75), kernel_size=(3,3), strides=1, padding='same', activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='same'))

    model.add(Conv2D(filters=int(filters*0.5), kernel_size=(3,3), strides=1, padding='same', activation='relu'))
    model.add(MaxPool2D(pool_size=(2,2), strides=2, padding='same'))

    model.add(Flatten())

    model.add(Dense(units=neurons, activation='relu'))
    model.add(Dropout(dropout))

    model.add(Dense(units=10, activation='softmax'))
    
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model

````

````python
from keras.callbacks import EarlyStopping

es = EarlyStopping(patience=10, min_delta=0.001, restore_best_weights= True) # Add early stopping

# Train model without fine tuning
model = create_model()
history = model.fit(train_generator,
                    epochs=1000,
                    validation_data=val_generator,
                    steps_per_epoch=len(train_data) // 128,
                    validation_steps=len(val_data) // 128,
                    callbacks=[es],
                    )
````

````python
from sklearn.metrics import classification_report

y_true = np.array([data[1] for data in test_data])
X_test = np.array([data[0] for data in test_data])

# evaluate model
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred , axis = 1) 

report = classification_report(y_true, y_pred)

print(report)
````

````python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

y_train = np.array([data[1] for data in train_data])
X_train = np.array([data[0] for data in train_data])

model = KerasClassifier(build_fn=create_model, verbose=0)

param_grid = {'filters': [25, 50, 75],
              'neurons': [128, 256, 512],
              'dropout': [0.0, 0.1, 0.2, 0.3]}

grid = GridSearchCV(estimator=model, param_grid=param_grid)

# Fine tune model
grid_result = grid.fit(X_train, y_train)

print("Best: %f using %s" % (grid_result.best_score_, 
                             grid_result.best_params_))
````

````python
# data augmentation datagen was taken from https://www.kaggle.com/code/codeblogger/convolutional-neural-network-cnn-tutorial

train_datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each input by its std
        zca_whitening=False,  # dimesion reduction
        rotation_range=5,  # randomly rotate images in the range 5 degrees
        zoom_range = 0.1, # Randomly zoom image 10%
        width_shift_range=0.1,  # randomly shift images horizontally 10%
        height_shift_range=0.1,  # randomly shift images vertically 10%
        horizontal_flip=False,  # randomly flip images
        vertical_flip=False)  # randomly flip images


# create new generator for training with usage of data augmentation
train_generator = train_datagen.flow(
        x=np.array([data[0] for data in train_data]),
        y=to_categorical(np.array([data[1] for data in train_data]), num_classes=10),
        batch_size=128
)
````

````python
from keras.callbacks import EarlyStopping


es = EarlyStopping(patience=10, min_delta=0.001, restore_best_weights= True) # Add early stopping

best_params = grid.best_params_

# Create a new model with the fine tuned hyperparameters
best_model = create_model(filters=best_params["filters"], 
                          neurons=best_params["neurons"], 
                          dropout=best_params["dropout"])

# Train fine tuned model
best_model.fit(train_generator,
                epochs=1000,
                validation_data=val_generator,
                steps_per_epoch=len(train_data) // 128,
                validation_steps=len(val_data) // 128,
                callbacks=[es])
````

````python
y_pred = best_model.predict(X_test)
y_pred = np.argmax(y_pred , axis = 1) 

report = classification_report(y_true, y_pred)

print(report)
````

````python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("loan_data.csv")

X = df.drop("credit.policy", axis=1)
X = pd.get_dummies(X, drop_first=True)
y = df["credit.policy"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123, stratify=y)
````

Decision tree:

````python
from sklearn.tree import DecisionTreeClassifier

# model without fine tuning with random hyperparameters
dt_no_tuning = DecisionTreeClassifier(
    max_depth=5, min_samples_leaf=1, min_samples_split=2)
dt_no_tuning.fit(X_train, y_train)

````

````python

from sklearn.model_selection import GridSearchCV

param_grid = {
    "max_depth": [5, 10, 15, 20, 30],
    "min_samples_split": [2, 5, 10, 15, 20, 30],
    "min_samples_leaf": [1, 2, 4, 8]
}

dt = DecisionTreeClassifier(random_state=123)

grid_search = GridSearchCV(estimator=dt, param_grid=param_grid)
grid_search.fit(X_train, y_train)

print("Best Hyperparameters:", grid_search.best_params_)
````

````python
best_params = grid_search.best_params_

# create model with fine tuned hyperparameters
dt = DecisionTreeClassifier(max_depth=best_params["max_depth"],
                            min_samples_leaf=best_params["min_samples_split"],
                            min_samples_split=best_params["min_samples_leaf"])

dt.fit(X_train, y_train)
````

    - Cons: 
       - Because ensemble methods call for training numerous models, they can be computationally expensive.
       - To combine the results from the various models, ensemble techniques need more complicated algorithms.
Ensemble techniques may not always lead to better model performance, especially if the individual models are not built well.
4. Briefly explain 2 types of boosting methods and 2 types of bagging methods.
Which of these categories does RF fall under?
    - Boosting methods:
      - Adaptive boosting 
        - uses decision trees, which typically contain a single split known as a decision stump, to represent weak learners. The primary decision stump in AdaBoost consists of observations with comparable weights.
      - Gradient Boosting 
        - increases the accuracy of the model by progressively adding predictors to the ensemble, whereby earlier predictors correct later ones.
    - Bagging methods:
      - Bootstrapping
        - is a sampling method where samples are drawn using the replacement process from the entire population (set). The sampling with replacement technique aids in the randomization of the selection process.
      - Aggregation 
        - is done to incorporate all possible outcomes of the prediction and randomize the outcome. Without aggregation, predictions will not be accurate because all outcomes are not put into consideration. 
        
    As it produces multiple decision trees using various subsets of the training data and then combines the predictions of all the trees it creates, Random Forest comes under the category of bagging methods.
